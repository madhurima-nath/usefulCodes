## Convert `.dat` file into dataframe

```python
import pandas as pd
import time
```

```python
# load entire contents of dat file into lust if binary objects

with open('foo.dat', 'rb') as file:
    text = file.readlines()
```

```python
# this decode the binary file as utf-8

text[0].decode()
```

```python
# check the position of the delimiter

# as an example for 1st entry
spchar = text[1].decode()[0]

for i in range(len(text)):
    if spchar != text[i].decode()[0]:
        print('delimiter position inconsistent')

print('delimiter position is different for items in the list')
```

```python
# check if all items in list have same length

list_length = len(text[0].decode().split(text[0].decode()[0]))

for i in range(len(text)):
    if list_length != len(text[i].decode().split(text[i].decode()[0])):
        print('length is not same for', i, '-th item')

print('length of all items are same')
```

To parse this file:
- decode binary contents
- split string at special characters as delimiter - only if position is the same
- run a loop through all items in list to get contents - only if format and length of all items are same

```python
stTime = time.time()

# define empty dict
d = {}

# declare delimiter
delimiter = text[0].decode()[0]     #0 to be replaced by position of delimiter

# run loop to get dict containing filename, label
for i in range(len(text)):
    # empty list to store filename, label for each row
    temp = []
    
    # include try-except to make sure code doesn't error out
    try:
        for j, k in enumerate(text[i].split(delimiter)):
            # if information required is only at odd indices
            if j%2 != 0:
                if '/' not in k:
                    temp.append(k)
        d[i] = temp
    except:
        pass

# keep track of time taken to complete the process
endTime = time.time()
tTime = (endTime - stTime)

print(f'Time taken to convert dat file contents into dict: {tTime:.2f} seconds')
print('process completed')
```

```python
# convert to dataframe

df = pd.DataFrame.from_dict(d, orient = 'index', columnd=['ID', 'Label'])
```
